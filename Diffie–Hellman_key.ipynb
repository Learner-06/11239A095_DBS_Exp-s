```python
def power(base, exp, mod):
    return pow(base, exp, mod)  # built-in fast modular exponentiation

def diffie_hellman():
    n = int(input("Enter prime number (n): "))
    g = int(input("Enter primitive root (g): "))

    x = int(input("Enter private key for Person 1: "))
    y = int(input("Enter private key for Person 2: "))

    a = power(g, x, n)  # public key of Person 1
    b = power(g, y, n)  # public key of Person 2

    key1 = power(b, x, n)  # shared secret for Person 1
    key2 = power(a, y, n)  # shared secret for Person 2

    print(f"\nKey for the first person: {key1}")
    print(f"Key for the second person: {key2}")

diffie_hellman()
```
