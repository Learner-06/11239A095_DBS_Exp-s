```python
import hashlib

def md5_hash():
    print("\t MD5 ENCRYPTION ALGORITHM IN PYTHON \n")

    msg = input("Enter the string to be encrypted using MD5:\n\t")
    result = hashlib.md5(msg.encode()).hexdigest()

    print("\nThe MD5 code for the input string is:")
    print(f"\t= 0x{result}")
    print("\n\tMD5 Encryption Successfully Completed!!!\n")

md5_hash()
```
