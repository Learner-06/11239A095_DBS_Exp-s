```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

def rsa():
    p = int(input("Enter first prime: "))
    q = int(input("Enter second prime: "))
    n = p * q
    phi = (p - 1) * (q - 1)

    # choose e
    e = next(i for i in range(2, phi) if gcd(i, phi) == 1)

    # find d
    d = next(i for i in range(2, phi) if (i * e) % phi == 1)

    print(f"\nPublic key: ({e}, {n})")
    print(f"Private key: ({d}, {n})")

    msg = input("\nEnter message (lowercase): ")
    enc = [pow(ord(c) - 96, e, n) for c in msg]
    print("Encrypted:", enc)

    dec = ''.join(chr(pow(c, d, n) + 96) for c in enc)
    print("Decrypted:", dec)

rsa()
```
