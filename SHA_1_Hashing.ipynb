```python
import hashlib

def bytes_to_hex(b):
    return ''.join(f'{byte:02X}' for byte in b)

def sha1():
    print("Message digest object info:")
    md = hashlib.sha1()
    print(f" Algorithm = {md.name}")
    print(f" Provider = {hashlib.__name__}")
    print(f" ToString = {md}\n")

    # Test inputs (same as original)
    for text in ["", "abc", "abcdefghijklmnopqrstuvwxyz"]:
        md = hashlib.sha1()
        md.update(text.encode())
        output = md.digest()
        print(f"SHA1(\"{text}\") = {bytes_to_hex(output)}\n")

if __name__ == "__main__":
    sha1()
```
